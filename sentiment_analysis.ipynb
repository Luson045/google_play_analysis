```python
df=df.dropna(subset=['Rating'])
for columns in df.columns:
  df[columns].fillna(df[columns].mode()[0],inplace=True)
df.drop_duplicates(inplace=True)
df=df[df['Rating']<=5]

df_rev.dropna(subset=['Translated_Review'],inplace=True)
```

```python
df.isnull().sum()
```

```python
#transformation
#converting installs to numeric
df['Installs']=df['Installs'].str.replace(',','').str.replace('+','').astype(int)
df.head()
```

```python
df.info()
```

```python
df['Price']=df['Price'].str.replace('$','').astype(float)
df.info()
```

```python
data=pd.merge(df,df_rev,on='App', how='inner')
```

```python
data.head()
```

```python
#data transformation
```

```python
#transforming sizes to MB
def convertsize(size):
  if 'M' in size:
    return float(size.replace('M',''))
  elif 'k' in size.lower():
    return float(size.lower().replace('k',''))//1024
  else:
    return np.nan
df['Size']=df['Size'].apply(convertsize)
```

```python
df.head()
```

```python
#logarithmic transformation
data['Reviews']=data['Reviews'].astype(int)
data['Log_Installs']=np.log(data['Installs'])
data['Log_Reviews']=np.log(data['Reviews'])
data.head()
```

```python
def rating_group(rating):
  if rating >=4:
    return 'Top rated app'
  elif rating >=3:
    return 'Above average'
  elif rating >=2:
    return 'average'
  else:
    return 'below average'
  
data['Rating_Groups']=data['Rating'].apply(rating_group)
data.head()
```

```python
#revenue column
data['Revenue']=data['Price']*data['Installs']
```

```python
def remove_outliers(df, columns):
  df_clean = df.copy()
  for column in columns:
    Q1=df_clean[column].quantile(0.25)
    Q3=df_clean[column].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    df_clean =df_clean[
      (df_clean[column]>=lower_bound)&
      (df_clean[column]<=upper_bound)
    ]
  return df_clean
```

```python
months=['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec']
fig5= go.Figure()
category_colors={}
for count,i in enumerate(category_year_installs.keys()):
  if i not in category_colors:
    category_colors[i]=f'rgba({count*30 % 256},{(count*60) % 256},{(count*90) % 256},1)'
  #calculating growth
  growth_mask=[False]*12
  for j in range(1,12):
    if(category_year_installs[i][j]>=1.2*category_year_installs[i][j-1]):
      growth_mask[j]=True
  fig5.add_trace(
    go.Line(
      x=months,
      y=category_year_installs[i],
      name=i,
      legendgroup=i,
      line=dict(color=category_colors[i])
    )
  )
  for j in range(1,12):
    if growth_mask[j]:
      fig5.add_trace(
        go.Scatter(
          x=[months[j-1],months[j]],
          y=[category_year_installs[i][j-1],category_year_installs[i][j]],
          mode='lines',
          fill='tozeroy',
          legendgroup=i,
          fillcolor=category_colors[i].replace(",1)", ",0.3)"),
          line=dict(width=0),
          showlegend=False
        )
      )
fig5.update_layout(
  title="Yearly Install Trends for each category",
  title_x=0.5,
  xaxis_title="Months",
  yaxis_title="Number of Installs",
  template="plotly_white"
)
save_plot_as_html2(fig5,"Year Wise Installs for each category.html","From the graph, we can see that a major amount of increase in installs for each category is seen betweek the month of JUNE-JULY, this may be due to the summer breaks, when people stay at home and are more likely to find ways to pass time.",timed=4)
```
